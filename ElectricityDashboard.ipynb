```python
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV file using numpy.genfromtxt
data_texas = np.genfromtxt(
    'Texas/Optimised/Texas_electricity_usage_2020.csv',        
    delimiter=';',          
    dtype=None,            
    encoding='utf-8',       
    names=True              
)

print(data_texas.dtype.names)
```

## Extracting Coloumns form np Array

```python
    

cities = data_texas['\ufeffCity']
energy_consump = data_texas['Value_kWh']

# Sorting the top 5 
sorted_indices = np.argsort(energy_consump)[::-1]
top_5_indices = sorted_indices[:5]  

# Extract top 5 cities and populations
top_5_cities = cities[top_5_indices]
top_5_energy_consump = energy_consump[top_5_indices]


```

## Graph

```python
# Create the bar chart
fig, ax = plt.subplots()

ax.bar(top_5_cities, top_5_energy_consump, color = 'red')
ax.set_ylabel('Value(kWh)')
ax.set_title('Electricity Consumption Texas (2020)')
ax.set_xticks(range(len(top_5_cities)))
ax.set_yticks(range(len(top_5_energy_consump)))
ax.set_xticklabels(top_5_cities, rotation=45, ha='right')  
plt.tight_layout()
plt.show()
```
